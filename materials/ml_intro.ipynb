```python
%matplotlib inline

import pathlib

import numpy as np
import matplotlib.pyplot as plt
```

```python
rng = np.random.default_rng(1)
num_points = 8
std = 0.3

x = rng.random((num_points,)) * 2 * np.pi
y = np.sin(x) + std * rng.normal(size=(num_points,))
z = np.linspace(0, 2 * np.pi, 500)
```

```python
for i in range(3):
    fig, ax = plt.subplots(1, 1, figsize=(5, 3))

    ax.scatter(x, y, color='C0')
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set(xlim=(0, 2*np.pi), ylim=(-1, 1))

    if i > 0:
        ax.axvline(4, -1, 1, color='C1', alpha=0.5)

    if i > 1:
        ax.plot(z, np.sin(z), color='C0', alpha=0.3, linewidth=2)

    fig.tight_layout()
    fig.savefig(pathlib.Path(f"~/Documents/ml_intro_{i}.pdf").expanduser())
```
